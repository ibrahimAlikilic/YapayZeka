## Sigmoid <br>
image.png

```python
def sigmoid(x):
  a=[]
  for i in x:
    a.append(1/(1+np.exp(-i)))
  return a
```

 Derin öğrenmede sigmoid fonksiyonu, bir girdi değerini 0 ile 1 arasında bir değere dönüştürmek için kullanılır. Genellikle sınıflandırma problemlerinde, özellikle iki sınıflı sınıflandırma (binary classification) işlemlerinde tercih edilir.<br>
a.append(1 / (1 + np.exp(-i))):: Burada her bir i değeri için sigmoid fonksiyonu uygulanır ve sonuç a listesine eklenir.

np.exp(-i): Bu ifade, negatif i değerinin üstel fonksiyonunu alır. exp(-i), e tabanında üssel bir işlemdir.
1 / (1 + np.exp(-i)): Bu, sigmoid fonksiyonunun kendisidir. Matematiksel olarak:<br>
image.png<br>
Bu formül, sonucu 0 ile 1 arasında bir değere dönüştürür.<br>
### Sigmoid Fonksiyonunun Özellikleri:
* Çıktı aralığı: (0, 1)
* Küçük negatif değerler için (örneğin, -10) sonucu 0'a çok yakın bir değer olur.
* Büyük pozitif değerler için (örneğin, 10) sonucu 1'e çok yakın bir değer olur.
* Orta değerler (örneğin, 0 civarında) için sonucu 0.5 civarındadır.<br>
Bu fonksiyonun bir alternatifi olarak, NumPy'nin yerleşik sigmoid fonksiyonu daha optimize edilmiş bir performans sağlar. np.vectorize() ile bu döngüyü tek satırda da yapabilirsin, ancak bu hali de manuel hesaplamayı gösterdiği için öğreticidir.

Keras'ta sigmoid fonksiyonu genellikle aktivasyon fonksiyonu olarak katmanlarda activation='sigmoid' şeklinde kullanılır.

## Hiperbolik Tanjant <br>
image.png


```python
def tanh(x, derivative=False):
    if (derivative == True):
        return (1 - (x ** 2))
    return np.tanh(x)
```

Bu fonksiyon, hiperbolik tanjant (tanh) fonksiyonunu hesaplıyor ve isteğe bağlı olarak bu fonksiyonun türevini de döndürebiliyor. Tanh fonksiyonu, sinir ağlarında aktivasyon fonksiyonu olarak sigmoid gibi yaygın bir şekilde kullanılır, ancak çıktı aralığı (-1, 1) arasında olduğundan, özellikle zıt kutuplu çıktıların önemli olduğu yerlerde tercih edilir.Bu fonksiyon, hiperbolik tanjant (tanh) fonksiyonunu hesaplıyor ve isteğe bağlı olarak bu fonksiyonun türevini de döndürebiliyor. Tanh fonksiyonu, sinir ağlarında aktivasyon fonksiyonu olarak sigmoid gibi yaygın bir şekilde kullanılır, ancak çıktı aralığı (-1, 1) arasında olduğundan, özellikle zıt kutuplu çıktıların önemli olduğu yerlerde tercih edilir.<br>
* derivative=False: Türev hesaplanıp hesaplanmayacağını belirleyen bir parametre.
### derivative=True =>
Tanh fonksiyonunun türevi, 1 - tanh(x)^2 şeklinde hesaplanır. Türev hesaplandığında, sinir ağı güncellemeleri için bu bilgi kullanılır. Türevin formülü şu şekildedir:<br>
image.png <br>
Ancak burada doğrudan tanh(x) yerine, kullanıcı zaten tanh(x) hesaplamış ve bunu girdi olarak veriyor gibi düşünülerek, x karesi alınıyor. Sonuç olarak, 1 - (x ** 2) ifadesi döndürülüyor.
### return (1 - (x ** 2)):
Burada, x zaten tanh fonksiyonunun sonucu olarak düşünülüyor ve onun karesi alınıp, 1'den çıkarılıyor.
### return np.tanh(x):
* Eğer türev hesaplanmıyorsa (yani derivative=False ise), NumPy'nin np.tanh(x) fonksiyonunu kullanarak girdinin hiperbolik tanjantını hesaplayıp döndürüyor.<br>
* np.tanh(x), ifadesi, NumPy kütüphanesinin bir fonksiyonu olan tanh (hiperbolik tanjant) fonksiyonunu çağırır ve girdi değeri x'e uygulanır. Hiperbolik tanjant, trigonometrik bir fonksiyon olan tanjantın hiperbolik versiyonudur ve matematiksel olarak şu şekilde tanımlanır:<br>
image.png<br>
Bu fonksiyon, giriş değerini -1 ile 1 arasında bir değere sıkıştırır. Pozitif girişler için çıktı 1'e, negatif girişler için çıktı -1'e yaklaşır.<br><br>
### Tanh Fonksiyonunun Özellikleri:
* Çıktı aralığı: (-1, 1)
* Pozitif girişler için çıktı 1'e yaklaşır.
* Negatif girişler için çıktı -1'e yaklaşır.
* Girdi değeri 0'a yakın olduğunda, çıktı da 0 civarındadır.
* Sigmoid fonksiyonundan farklı olarak, merkezde simetriktir ve çıktı negatif değerlere de ulaşabilir. Bu, özellikle gradient descent gibi öğrenme algoritmalarında, zıt yöndeki hataları daha iyi işleyebilmesini sağlar.

# ReLu
image.png

```python
def re(x):
  b=[]
  for i in x:
    if i<0:
      b.append(0)
    else:
      b.append(i)
  return b
```

# Leaky ReLu
image.png

```python
def lr(x):
  b=[]
  for i in x:
    if i<0:
      b.append(i/10)
    else:
      b.append(i)
  return b
```

# Swish
image.png <br>
## swish=sig*x

```python
# Grafiklerin ooluşturulması için aralıkların tanımlanması
x=np.arange(-3,3,0.1)
sig=sigmoid(x)
tanh=tanh(x)
relu=re(x)
leaky_relu=lr(x)
swish=sig*x
```

```python
# fonksiyonların ekrana çizdirilmesi
line1,=plt.plot(x,sig,label="sigmoid")
line2,=plt.plot(x,tanh,label="Tanh")
line3,=plt.plot(x,relu,label="Relu")
line4,=plt.plot(x,leaky_relu,label="Leaky_Relu")
line_5,=plt.plot(x,swish,label="Swish")
plt.legend(handles=[line1,line2,line3,line4,line_5])
plt.axhline(y=0,color="k")
plt.axvline(x=0,color="k")
plt.show()
```

### 1) plt.axhline(y=0, color="k")
* plt.axhline: Bu fonksiyon, grafiğe yatay bir çizgi ekler.
* y=0: Bu parametre, çizginin y ekseninde nerede çizileceğini belirtir. Burada, * yatay çizgi y = 0 noktasında çizilir (yani x ekseni boyunca çizilir).
* color="k": Bu, çizginin rengini belirtir. "k" siyah rengi temsil eder.<br>(Matplotlib'de renk kodları için kısayollar vardır; "k" siyah için kullanılır).
Bu satır, grafikte x ekseni boyunca siyah renkte yatay bir çizgi çizer.

### 2) plt.axvline(x=0, color="k")
* plt.axvline: Bu fonksiyon, grafiğe dikey bir çizgi ekler.
* x=0: Bu parametre, çizginin x ekseninde nerede çizileceğini belirtir. Burada, * dikey çizgi x = 0 noktasında çizilir (yani y ekseni boyunca çizilir).
* color="k": Bu da çizginin rengini siyah olarak ayarlar.
